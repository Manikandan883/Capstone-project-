#### Decoding the categoric columns

```python
df1= data.copy()
```

```python
# Decoding Gender
df1['Gender'] = df1['Gender'].replace({0: 'Female', 1: 'Male'}).astype('object')

# Decoding Medication Status
df1['Medication_Status'] = df1['Medication_Status'].replace({0: 'No Medication', 1: 'On Medication'}).astype('object')

# Decoding Multi class label
df1['Multi_Class_Label'] = df1['Multi_Class_Label'].replace({
    0: "Normal (Non-Seizure)", 
    1: "Pre-Seizure (Pre-Ictal Phase)", 
    2: "Seizure (Ictal Phase)", 
    3: "Post-Seizure (Post-Ictal Phase)"
}).astype('object')

# Decoding Seizure type label
df1['Seizure_Type_Label'] = df1['Seizure_Type_Label'].replace({
    0: "Normal", 
    1: "Generalized Seizure", 
    2: "Focal Seizure"
}).astype('object')
```

#### Binning Age

```python
def get_age_group(age):
    if age <= 4:
        return 'Infant(<= 4)'
    elif age <= 11:
        return 'Child(<= 11)'
    elif age <=17:
        return 'Teen(<=17)'
    elif age <=25:
        return 'YoungAdult(<=25)'
    elif age <= 44:
        return ' Adult(<= 44)'
    elif age<=64:
        return 'Middle_Aged_Adult(<=64)'
    else:
        return "Seinor(65+)"


df1['Age Group'] = df1['Age'].apply(get_age_group)
```

```python
df1.drop(['Age'], axis=1 ,inplace= True)
df1
```

#### PCA

```python
df2.drop(columns = 'Seizure_Type_Label' , axis=1, inplace= True)
```

```python
df2.drop(columns = 'Multi_Class_Label' , axis=1, inplace= True)
```

```python
df3 = df2.copy()
df3
```

```python
df4 = df3.copy()
df4.shape
```

```python
cov_matrix=np.cov(df4.T) 
cov_matrix
```

```python
eigval,eigvect=np.linalg.eig(cov_matrix) 
```

```python
eigval
```

```python
eigvect
```

```python
tot = sum(eigval)
tot
```

```python
var_exp = [(i/tot)*100 for i in sorted(eigval, reverse=True)]
var_exp
```

```python
cum_exp_var = np.round(np.cumsum(var_exp),3)
cum_exp_var
```

```python
from sklearn.decomposition import PCA 
mypca=PCA(n_components=0.95) 
pca1=mypca.fit_transform(df4) 
pca1_df=pd.DataFrame(pca1) 
pca1_df.head()  
```
